```python
import pandas as pd
from pandas import Grouper
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('data.csv')
```

```python
#Task1.1 - Find songs with value for energy column greater than 0.4 and display values for song_title column.
f1=df.sort_values('energy')['song_title']
f1
```

```python
#Task1.2 - For songs with duration duration_ms value less than 300000 ms (300 s), find Top 10 “fastest” songs( songs with highest values for tempo column ) and display values for song_title, artistcolumns.
f2=df[df['duration_ms']<300000]
f3=f2.sort_values('tempo')[['song_title','artist']].tail(10)
f3
```

```python
#Task1.3 - Find songs with value for artist column = Drake and value for energy column greater than 0.5 and display values for song_title column.
f4=df[df['artist']=='Drake'][df['energy']>0.5]
f5=f4['song_title']
f5
```

```python
#Task2.1 - Create a dataframe of the songs that rank among top 10 for the instrumentalness metric. This dataframe should only contain the instrumentalness, song title and artist columns
f6=df.sort_values('instrumentalness',ascending=False)[['song_title','artist','instrumentalness']].head(10)
f6
```

```python
#Task2.2 - Create a dataframe of the average tempo for each of the top 10 most occurring artists.
f7=df['artist'].value_counts()[:10].index.tolist()
f8=df.groupby('artist')['tempo'].mean().to_dict()
f9={}
for i in f7:
   f9[i]=f8[i]
f10=pd.DataFrame(list(f9.items()),columns = ['Artist', 'Average Tempo'])
f10
```
